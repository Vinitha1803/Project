```python
def remove_company_suffix(org):
    if pd.notna(org):
        suffixes = ["Inc.", "Ltd.", "Corp.", "LLC", "Co.", "AG", "GmbH", "BV", "S.A.", "Pvt. Ltd.", "LLP"]
        for suffix in suffixes:
            org = org.replace(suffix, "").strip()
    return org
```

```python
f_cleaned['Organization']
```

```python
f_cleaned['Organization'] = f_cleaned['Organization'].apply(remove_company_suffix)
f_cleaned['Organization']
```

```python
from opencage.geocoder import OpenCageGeocode
import folium
```

```python
def geocode_company(org):
    # Replace "your_opencage_api_key" with your actual OpenCage Geocoding API key
    api_key = "feecb3f7edc84718a351e579606425e8"
    geocoder = OpenCageGeocode(api_key)

    try:
        result = geocoder.geocode(org)
        if result and len(result):
            latitude = result[0]['geometry']['lat']
            longitude = result[0]['geometry']['lng']
            return latitude, longitude
    except Exception as e:
        print(f"Error geocoding {org}: {e}")

    return None, None
```

```python
f_cleaned['Latitude'], f_cleaned['Longitude'] = zip(*f_cleaned['Organization'].apply(geocode_company))
print(f_cleaned)
```

```python
map_center = [f_cleaned['Latitude'].mean(), f_cleaned['Longitude'].mean()]
connections_map = folium.Map(location=map_center, zoom_start=10)

for _, row in f_cleaned.iterrows():
    if pd.notna(row['Latitude']) and pd.notna(row['Longitude']):
        folium.Marker(
            location=[row['Latitude'], row['Longitude']],
            popup=row['First Name'],
        ).add_to(connections_map)
connections_map.save("connections_map.html")
```
